```python
# Step 2: Import necessary library
from transformers import pipeline

# Step 3: Load the zero-shot-classification model
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
# Step 4: Define the code snippet to analyze
code_snippet = "import os; while;"

# Step 5: Define candidate labels
labels = ["malicious", "safe"]

# Step 6: Run classification
result = classifier(code_snippet, labels)

# Step 7: Display result
import pandas as pd
pd.DataFrame({
    "Label": result["labels"],
    "Score": result["scores"]
})
```
